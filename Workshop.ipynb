```r
library(pheatmap)
library(RColorBrewer)
```

```r
mRNA = read.delim("data_control_61_disease_45.txt",
                  sep='\t',
                  row.names = 1)
mRNA
```

```r
names(mRNA)
```

```r
group_df = data.frame(Groups=rep(c("Control", "PTB"),
                                 c(61,45)))

rownames(group_df) <- colnames(mRNA)
```

```r
p<-pheatmap(mRNA,
         cluster_cols = FALSE, scale = 'row',
         annotation_col = group_df,
         show_colnames = FALSE,
         colorRampPalette(c("navy", "white", "#FF1493"))(75),
         gaps_col = cumsum(c(61,45)))
```

```r
png(file="heat_mapxxxx.png", width=18, height=4, units="in", res=300)
p
dev.off()
```

```r
M<-cor(t(mRNA), method='pearson')
head(round(M,2))
```
